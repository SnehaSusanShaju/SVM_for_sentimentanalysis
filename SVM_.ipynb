```python
import re
```

```python
def removing():
  new_text=[]
  for i in df['tweet']:
    text=re.sub(r'https?://\S+|[^0-9A-Za-z t]|@\w+'," ",i)
    new_text.append(text)
  return new_text
```

```python
r=removing()
```

```python
r
```

```python
new_df=pd.DataFrame(r,columns=['tweet'])
new_df['label']=df.label
```

```python
new_df.head()
```

```python
new_df["tweet"].str.lower()
```

```python
X=new_df["tweet"]
y=new_df["label"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42) 
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# Create feature vectors
vectoriser = TfidfVectorizer(min_df = 5,
                             max_df = 0.8,
                             sublinear_tf = True,
                             use_idf = True)
train_v = vectoriser.fit_transform(X_train)
test_v= vectoriser.transform(X_test)
     
     
```

```python
from sklearn.svm import SVC  
classifier = SVC(kernel='linear') 
classifier.fit(train_v, y_train)
```

```python
pred = classifier.predict(test_v)
```

```python

from sklearn.metrics import classification_report
print(classification_report(y_test, pred))
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred,y_test)
acc
```
